```python
titanic_df.columns
```

```python
titanic_df.head(10)
```

```python
titanic_df.replace({'Gender': {'male': 0}}, inplace = True)
titanic_df.replace({'Gender': {'female': 1}}, inplace = True)
titanic_df.head(10)
```

```python
import seaborn as sns
sns.boxplot(y = titanic_df['Fare'] )
```

```python
titanic_df.drop(titanic_df[titanic_df.Fare >= 500].index, inplace = True)
sns.boxplot(y = titanic_df['Fare'] )
```

```python
titanic_df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
titanic_df[['Fare']] = scaler.fit_transform(titanic_df[['Fare']])
titanic_df.describe()
```

### 1. Data cleaning -  

```python
# every feature not null
employee.info()
```

```python
#    
list = employee.columns
list
```

```python
employee[employee.duplicated(subset=list)].head()
```

```python
employee.drop_duplicates(inplace = True)
employee
```

```python
#  
employee[employee.duplicated(subset=list)].head()
```

### 2. Data transformation

#### 2-1)  (feature rename)

```python
employee.rename(columns={'ExperienceInCurrentDomain': 'Experience',
                         'Education': 'Edu',
                         'LeaveOrNot': 'Leave',
                         'JoiningYear': 'Join'}, inplace=True)
employee.columns
```

#### 2-2)  (data encoding)

```python
print(employee['Edu'].unique())
print(employee['City'].unique())
print(employee['Gender'].unique())
print(employee['EverBenched'].unique())
```

```python
employee.replace({'Edu':{'Bachelors' : 0}}, inplace = True)
employee.replace({'Edu':{'Masters' : 1}}, inplace = True)
employee.replace({'Edu':{'PHD' : 2}}, inplace = True)
employee.replace({'City':{'Bangalore' : 0}}, inplace = True)
employee.replace({'City':{'Pune' : 1}}, inplace = True)
employee.replace({'City':{'New Delhi' : 2}}, inplace = True)
employee.replace({'Gender':{'Male' : 0}}, inplace = True)
employee.replace({'Gender':{'Female' : 1}}, inplace = True)
employee.replace({'EverBenched':{'No' : 0}}, inplace = True)
employee.replace({'EverBenched':{'Yes' : 1}}, inplace = True)
employee.head()
```

```python
employee.reset_index(drop=True, inplace=True)
```

### 3. Data normalization - MinMax Scaling

```python
from sklearn.preprocessing import MinMaxScaler
y_employee = employee['Leave']
X_employee = employee.drop(columns='Leave')

scaler = MinMaxScaler()
X_employee = pd.DataFrame(scaler.fit_transform(X_employee), columns=X_employee.columns)
X_employee
```

```python
employee_df = pd.concat([X_employee, y_employee], axis=1)
employee_df
```
